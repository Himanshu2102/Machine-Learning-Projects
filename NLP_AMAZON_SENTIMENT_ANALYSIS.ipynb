```python
import string
string.punctuation
```

```python
from nltk.corpus import stopwords #natural language tool kit
```

```python
def clean_text(X):
    clean_X=[charachter for charachter in X if charachter not in string.punctuation]
    clean_X_join = ''.join(clean_X)
    cleaned_text = [word for word in clean_X_join.split() if word.lower() not in stopwords.words('english')]
    return(cleaned_text)
```

```python
#check for function working
df_amazon['verified_reviews'][3]
```
